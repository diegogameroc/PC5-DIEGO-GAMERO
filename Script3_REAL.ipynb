# EJERCICIO
Crear un dataframe que tenga como columnas:
La media artimetica de los minutos de recreo
MOda de los minutos de recreo
Numero de colegios

Para cada Departamento

```python
# Construir listar para cada uno de los items requeridos
Lista_medias = list()
lista_modas = list()
lista_numCol = list()

for dpto in Inicial.departamento.unique():
  data=Inicial.loc[Inicial.departamento==dpto, "p2_1_02_006_2_1"]
  Lista_medias.append(data.mean())
  lista_modas.append(data.mode()[0])
  lista_numCol.append(data.shape[0])
```

```python
# Creacion de un diccionario dpto_recreo
dpto_recreo=pd.DataFrame.from_dict({}.fromkeys(Inicial.departamento.unique()),orient ="index")
dpto_recreo["Promedio"]=Lista_medias
dpto_recreo["Moda"]=lista_modas
dpto_recreo["Numero de colegios"]=lista_numCol
dpto_recreo = dpto_recreo.drop([0],axis=1)
dpto_recreo
```

```python
# Guardemos en disco duro los calculos que acabamos de obtener en la celda anterior
dpto_recreo.to_csv("dpto_media_moda_numColegios.csv")
```
